### `order_products` data frame

Now let's fill in missing values from the order_products table.

````python
# Display rows where the add_to_cart_order column has missing values
order_products[order_products['add_to_cart_order'].isna()] # Same logic as previous two sets of operations 
````

````python
# Use .min() and .max() to find the minimum and maximum values for this column.
print(order_products['add_to_cart_order'].min())
print(order_products['add_to_cart_order'].max()) # print() statements added to display both. Run order prioritizes last line without them
````

````python
# Save all order IDs with at least one missing value in 'add_to_cart_order'
missing_rows = order_products[order_products['add_to_cart_order'].isna()] # Filter NaN rows
missing_order_ids = missing_rows['order_id'].unique() # Extract unique order_id numbers as a data subset from missing_rows
````

````python
# Do all orders with missing values have more than 64 products?

# Use .isin() to filter order_products to those order_ids, then groupby to associate order_id with related product_id numbers, and count.
missing_order_counts = order_products[order_products['order_id'].isin(missing_order_ids)].groupby('order_id')['product_id'].count()
print(f"Minimum products in orders with missing values: {missing_order_counts.min()}")

# Check if all counts > 64 with .min() > 64       
print(f"Are ALL orders with missing values > 64 products? {missing_order_counts.min() > 64}") # Conditional to check

# Get the order IDs that have missing add_to_cart_order values
missing_order_ids = order_products[order_products['add_to_cart_order'].isna()]['order_id'].unique()
print(f"Number of orders with missing values: {len(missing_order_ids)}") # This is a numpy array. .count() doesn't work here.
````

````python
# Replace missing values with 999 and convert column to integer type
order_products['add_to_cart_order'].fillna(999, inplace=True)
order_products['add_to_cart_order'] = order_products['add_to_cart_order'].astype('int')
# NaN forces float64. .fillna() has to come before astype('int')

print(order_products[order_products['add_to_cart_order'].isna()]) # Checks out
````

For some reason, any item placed in the cart 65th or later has a missing value in the `'add_to_cart_order'` column. Maybe the data type of that column in the database could only hold integer values from 1 to 64. We've decided to replace the missing values with a code value, 999, that represents an unknown placed in cart order above 64. We also converted the column to integer data type. We just need to be careful to remember this if we perform calculations using this column during our analysis.

Other sensible code values we could've used are 0 or -1 because they don't show up elsewhere in the dataset and they don't have any real physical meaning for this variable.

Also note that, for orders with exactly 65 items, we could replace the missing value with 65. But we're going to neglect that for now since we can't determine the 65th item for all orders with 66 items or more.
